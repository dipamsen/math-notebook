```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sin = np.sin
cos = np.cos
id = lambda x: x
```

# $\sin{x}$ vs. $x$

```python
inp = np.linspace(-np.pi, np.pi, 100)
plt.plot(inp, sin(inp), label='sin(x)')
plt.plot(inp, (inp), label='x')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.axis('square')
plt.show()
```

- For $\sin{x}$, both the graphs only intersect at one point, $x=0$.
- In $x > 0$,        $\qquad x > \sin{x} \qquad \forall x$.
- In $x < 0$,        $\qquad x < \sin{x} \qquad \forall x$.

# $\cos{x}$ vs. $x$

```python
inp = np.linspace(-np.pi/2, np.pi/2, 100)
plt.plot(inp, cos(inp), label='cos(x)')
plt.plot(inp, (inp), label='x')
plt.grid(True)
plt.legend()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.axis('square')
plt.show()
```

- For $\cos{x}$, both the graphs intersect at a single non-trivial point.
- Let this intersection point be at $x = \lambda$. ($\lambda \approx 0.739$)
- In $x > \lambda$,        $\qquad x > \cos{x} \qquad \forall x$.
- In $x < \lambda$,        $\qquad x < \sin{x} \qquad \forall x$.
